```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
breast_data.shape
```

```python
breast_input = pd.DataFrame(breast_data)
breast_input.head()
```

```python
breast_labels = breast.target
breast_labels.shape
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
final_breast_data = np.concatenate([breast_data,labels], axis=1)
final_breast_data.shape
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
features = breast.feature_names
features
```

```python
features_labels = np.append(features, 'label')
```

```python
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign', inplace = True)
breast_dataset['label'].replace(1, 'Malignant', inplace = True)
breast_dataset.tail()
```

```python
breast_dataset.head()
```

```python
X = breast_dataset.iloc[:, 0:30].values
Y = breast_dataset.iloc[:, 30].values
```

```python
X.shape
```

```python
Y.shape
```

```python
print (range(25))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# Creating a "for" loop to see all the results of changing the number of components and scores, together; 

Accuracy = np.zeros(31)
Precision = np.zeros(31)
Recall = np.zeros(31)

for k in range(1,31):  
    pca = PCA(n_components = k)
    principalComponents = pca.fit_transform(X)
    
    X_train, X_test, Y_train, Y_test = train_test_split(principalComponents, Y, test_size = 0.2, random_state = 0)
    
    classifier = GaussianNB()
    classifier.fit(X_train, Y_train)
    Y_expected = Y_test
    Y_predic = classifier.predict(X_test)
    
    confusion_matrix = metrics.confusion_matrix(Y_expected, Y_predic)
    print (f"n_component = {k}")
    
    Accuracy[k] = metrics.accuracy_score(Y_expected, Y_predic)
    Precision[k] = metrics.precision_score(Y_expected, Y_predic, pos_label="Benign")
    Recall[k] = metrics.recall_score(Y_expected, Y_predic, pos_label="Benign")
    
    print('Training accuracy:',  classifier.score(X_train, Y_train))
    print('Test accuracy:',  classifier.score(X_test, Y_expected))
    print("\n")
    print(metrics.classification_report(Y_expected, Y_predic))
    print(metrics.confusion_matrix(Y_expected, Y_predic))
     
    print("\n\n")
```

```python
xaxis = range(1,32)
fig = plt.figure(dpi=300)
plt.ylim([0.7, 1.0])
plt.title('Model Performance', fontsize = 12)
plt.xlabel("Number of Principal Components (K)", fontsize = 9)
plt.ylabel("Accuracy, preceision, Recall", fontsize = 9)
plt.plot(xaxis, Accuracy, label="accuracy", color = 'red') 
plt.plot(xaxis, Precision, label="precision", color = 'green')
plt.plot(xaxis, Recall, label="recall", color = 'blue')
fig.legend(loc = "upper right")
plt.grid()
```
